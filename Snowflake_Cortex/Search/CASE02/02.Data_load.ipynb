## STEP02.Snowflake에 데이터 로드

Kaggle에서 다운로드한 파일을 저장할 스테이지를 생성

데이터 세트는 Snowsight 에서 업로드하거나 SQL을 사용하여 업로드 할수 있습니다.

### Snowsight 업로드하기      
1. Snowsight 에 로그인합니다.

2. 왼쪽 탐색 메뉴에서 **Data** 를 선택합니다.

3. **cortex_search_tutorial_db** 데이터베이스를 선택합니다.

4. 스키마 **public** 을 선택합니다.

5. **Stages** 를 선택하고 **books_data_stage** 를 선택합니다.

6. 오른쪽 상단에서 **+ Files** 버튼을 선택합니다.

7. UI에 파일을 끌어서 놓거나 **Browse** 를 선택하여 대화 창에서 파일을 선택합니다.

8. **Upload** 를 선택하여 **BooksDatasetClean.csv** 파일을 업로드합니다.

9. 파일 오른쪽에 있는 세 개의 점을 선택하고 **Load into table** 을 선택합니다.

10. 테이블 이름을 **BOOKS_DATASET_RAW** 로 지정하고 **Next** 를 선택합니다.

11. 데이터 로드 대화 상자의 왼쪽 패널에 있는 **Header** 메뉴에서 **First line contains header** 를 선택합니다.

12. 그런 다음 **Load** 를 선택합니다.
